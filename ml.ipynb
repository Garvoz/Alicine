```python
from sklearn.neighbors import NearestNeighbors

def mes_recommendations(movie_title, nb_films=5):
    if movie_title not in movies['Titre'].values:
        return "Le film demandé n'est pas dans la base de données."


    features = movies.drop(columns=['Titre', 'Genres', 'Date de sortie', 'Langue Originale', 'Résumé', 'ID', 'Lien du poster', 'Budget', 'Titres possibles', 'Nb de votes','Liste acteurs','Réalisateurs'])
    features['Décénnie'] *= 2.5
    features['Score de popularité'] *= 2
    features['Note moyenne'] *= 1.5
    features['Durée (minutes)'] *= 1


    knn = NearestNeighbors(n_neighbors=nb_films+1, metric='cosine')
    knn.fit(features)


    movie_index = movies[movies['Titre'] == movie_title].index[0]
    distances, indices = knn.kneighbors(features.iloc[movie_index].values.reshape(1, -1))
    recommended_movies = movies.iloc[indices[0][1:]]['Titre'].values
    recommended_movie_ids = movies.iloc[indices[0][1:]]['ID'].values

    # je commence la partie de recommendation de  films français ici 
    french_movies = movies[movies['Langue Originale'] == 'fr']
    filtered_movies = french_movies[~french_movies['Titre'].isin(recommended_movies)]

    filtered_features = filtered_movies.drop(columns=['Titre', 'Genres', 'Date de sortie', 'Langue Originale', 'Résumé', 'ID', 'Lien du poster', 'Budget', 'Titres possibles', 'Nb de votes','Liste acteurs','Réalisateurs'])
    features['Décénnie'] *= 2.5
    features['Score de popularité'] *= 2
    features['Note moyenne'] *= 1.5
    features['Durée (minutes)'] *= 1

    knn_filtered = NearestNeighbors(n_neighbors=nb_films, metric='cosine')
    knn_filtered.fit(filtered_features)

    french_distances, french_indices = knn_filtered.kneighbors(features.iloc[movie_index].values.reshape(1, -1))
    recommended_films_français = filtered_movies.iloc[french_indices[0]]['Titre'].values
    recommended_french_movie_ids = filtered_movies.iloc[french_indices[0]]['ID'].values
    print 

    return {
        "Recommandations générales": list(zip(recommended_movies, recommended_movie_ids)),
        "Recommandations en français": list(zip(recommended_films_français,recommended_french_movie_ids))
    }

resultat = mes_recommendations('Night School', 8)
```
